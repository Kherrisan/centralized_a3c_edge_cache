```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import pandas
import numpy as np
```

```python
colors={
    'random':'#383838',
    'lru':'#F03600',
    'lfu':'#36AB43',
    'a3c':'#1F539C'
}
labels = {'random':'Random','lru':'LRU','lfu':'LFU','a3c':'Proposed'}
markers={
    'random':'x',
    'lru':'o',
    'lfu':'v',
    'a3c':'s'
}
DPI = 100
```

```python
plt.figure(dpi=DPI)
ad_e = pandas.read_csv('./data/4_ad_e.csv')
ad_e =  ad_e.drop('Step', axis=1)
m_ad_e = ad_e.groupby(np.arange(len(ad_e))//100).mean()
handlers = []
for alg in m_ad_e.columns:
    plt.plot(m_ad_e[alg], colors[alg], lw=2.5,label=labels[alg])
plt.xlabel("Episode")
plt.ylabel("Average Delay(ms)")
plt.legend(loc=1)
plt.gca().xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, pos: "%d00"%(x,) if x!=0 else "0"))
plt.grid()
plt.savefig('./images/ad_e.jpg')
plt.show()
```
